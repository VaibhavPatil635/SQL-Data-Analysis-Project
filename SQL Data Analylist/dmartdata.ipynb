```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine(
    "mysql+pymysql://root:1111@localhost:3306/retail_db"
)
```

```python
query = "SELECT * FROM retail_raw"
df = pd.read_sql(query,engine)
```

```python
df.drop(columns=["Cancellation_Date"],inplace=True)
```

```python
# Datatype change
df["Customer_Age"] = df["Customer_Age"].astype("int64")
df["MRP"] = df["MRP"].astype("Float64")
df["Discount_Price"] = df["Discount_Price"].astype("float64")
df["Pin_Code"] = df["Pin_Code"].astype("int64")
df["Total_Order_Value"] = pd.to_numeric(df["Total_Order_Value"],errors="coerce")
df["No_of_Clicks"] = df["No_of_Clicks"].astype("int64")
df["Year"] = df["Year"].astype("int64")
df["Shipping_Charges"] = pd.to_numeric(df["Shipping_Charges"],errors="coerce")
df["Bill_Number"] = pd.to_numeric(df["Bill_Number"],errors="coerce")
df["Time_Spent_on_Website"] = df["Time_Spent_on_Website"].astype("float64")
df["Rating"] = pd.to_numeric(df["Rating"],errors="coerce")
```

```python
df.columns
```

```python
# Convert datetime formate
df["Delivery_Date"] = pd.to_datetime(df["Delivery_Date"],errors="coerce")
```

```python
df["Order_Date"] = pd.to_datetime(
    df["Order_Date"],
    dayfirst=True,
    errors="coerce"
)
```
